```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
#이번에는 사진 옆으로 단순하게 이어붙이기 위해 패치버전으로 출력


# 이미지 불러오기
img1 = cv2.imread('biimg1.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('biimg2.png', cv2.IMREAD_GRAYSCALE)

# 크기 통일하기
if img1.shape != img2.shape:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# and와 or연산 진행
bitwise_and = cv2.bitwise_and(img1, img2)
bitwise_or = cv2.bitwise_or(img1, img2)

# 결과 출력
print("원본이미지 두장, AND연산과 OR연산 결과 출력")
combined_logic = np.hstack((img1, img2, bitwise_and, bitwise_or))
cv2_imshow(combined_logic)
```

```python
# img1을 대상으로 NOT 연산 실행
bitwise_not = cv2.bitwise_not(img1)

print("원본과 NOT연산 결과 출력")
cv2_imshow(np.hstack((img1, bitwise_not)))
```
